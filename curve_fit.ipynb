```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
#membuat rng(random number generate/generator angka acak)
rng = np.random.RandomState(1)

#membuat data acak x antara 0 hingga 10
x = 10 * rng.rand(50)

#membuat data y sebagai fungsi sinus dengan noise
y = np.sin(x) + 0.1 * rng.randn(50)

#visualisasi data acak
plt.scatter(x, y, label="Data Acak")
plt.title("Scatter Plot Data Acak")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
#membuat model regresi linier
model = LinearRegression(fit_intercept=True)

#melatih model menggunakan data x dan y
model.fit(x[:, np.newaxis], y)

#menampilkan parameter model
print("Model slope:    ", model.coef_[0])   #kemiringan garis regresi
print("Model intercept:", model.intercept_) #intersep garis regresi
```

```python
#membuat data prediksi untuk x
xfit = np.linspace(0, 10, 1000) #1000 titik antara 0 hingga 10

#prediksi nilai y menggunakan model regresi
yfit = model.predict(xfit[:, np.newaxis])

#visualisasi data asli dan garis regresi
plt.scatter(x, y, label="Data Asli")                        #scatter plot data asli
plt.plot(xfit, yfit, color="red", label="Regresi Linier")   #garis regresi
plt.title("Regresi Linier Data")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
